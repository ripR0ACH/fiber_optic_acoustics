# Examples

The purpose of this package is to simulate the potential setups for our fiber optic acoustic detector system. To begin, it's best to start with the `fiber.py` package. The main class in this package is `fiber.FiberBundle(r, core = 2e-5, cladding = 0)`. This class is used to instantiate a fiber bundle. It takes three paramters: the radius of the whole fiber bundle, `r` (required); the radius of the core of each individual fiber, `core` (optional); the radius of the cladding of each individual fiber, `cladding`.

### Making a fiber bundle.
Let's begin with showing how to instantiate a fiber bundle.

```python
# import the FiberBundle class from the fiber.py class
from fiber import FiberBundle
# instantiate a FiberBundle object with a total radius of 
# of 5e-4 and a core radius of 1e-5 (measured in meters).
bundle = FiberBundle(5e-4, 1e-5, 5e-5) 
```

### Getting information about the bundle.

There are a number of different pieces of information about the fiber bundle that're saved into the object. These can be extracted from the bundle's attributes. 

##### Fiber core.

The fiber core is the part of an individual fiber that is sensitive to the light incident on it's surface. It's the second parameter passed into the bundles' instantiation above, but it can also be pass in as a named parameter called `core`. It is defined as a radius stored within the fiber bundles' attributes as `fiber.FiberBundle.core`.

```python
bundle.core
```

***
##### Fiber cladding.

The fiber bundle also has an attribute that defines the cladding on each individual fiber. By default, the cladding is set to zero, but it can be defined as the third paramter in the bundle instantiation or by the keyword `cladding`. This value is stored within the bundle object as `fiber.FiberBundle.cladding`.

```python
bundle.cladding
```

***

##### Fiber radius.

The fiber radius is the total radius of an individual fiber, i.e. the fiber core plus the cladding. This value is stored within the bundle object as `fiber.FiberBundle.fr`.

```python
bundle.fr
```

***
##### Fiber rings.

The radius of each "ring" of fibers that're filled within the fibers is held in `fiber.FiberBundle.fiber_rings`. These rings are used in the filling algorithm to maximize the number of fibers to be fit within the fiber bundle.

```python
bundle.fiber_rings
```

***
##### Fibers.

All of the fibers are stored within an attribute of the object called `fiber.FiberBundle.fibers`. This attribute is very bare bones. Don't expect to get very much information out of this attribute because most of the true information about the fibers is discretized and stored in other attributes on the object. This attribute is mainly used for the instantiation process of the object. This attribute is a list of lists of lists. That may be confusing, but the list contains lists. These lists contain one more layer of lists. The innermost list is four values structured as so `[theta, ring, ring number, fiber half]` where `theta` is the theta position of the individual fiber within the bundle, `ring` is the radius of the ring within which the fiber is contained, `ring number` is the index of the ring that the fiber is within in the bundle, and `fiber half` is the half of the fiber that the ring is on (top or bottom). The list surrounding the innermost list is the collection of all of the fibers contained within a single ring, and the list surrounding the lists of lists contains all of the rings of fibers. I feel that this will make sense more if you see it.

```python
bundle.fibers
```

***

##### Fiber centers.

Every fiber has a center, i.e. where the individual fiber's radius equals zero, and these values are stored in order of each fiber's instantiation sequentially. It follows from this that each fiber has an individual index. The centers of these fibers are stored in a list within the bundle object according to their index. The centers are stored within `fiber.FiberBundle.centers`. These centers are in the format of `[[x0, y0], [x1, y2], ...]` coordinates.

```python
bundle.centers
```

***

##### Fiber thetas.

The fiber thetas are the polar angular positions of each individual fiber. These are stored according to the same convention as the centers of the fiber. The unit of the angles are in radians. They can be accessed with the attribute `fiber.FiberBundle.thetas`.

```python
bundle.thetas
```

***

##### Fiber ring radius.

The fiber ring radius tells you the radius at which the fiber is located. This attribute is somewhat redundant, however it can be useful in specific cases where you want to know the radius at which a specific fiber is within the bundle. It is stored with the same convention as the other indexable attributes above, and it can be accessed with `fiber.FiberBundle.ring_radius`.

```python
bundle.ring_radius
```

***

##### Fiber ring number.

The fiber ring number tells you the number of the ring that the fiber is drawn in. The innermost ring is the zeroth ring and the rest of the rings are counted outward. This attribute can be accessed with `fiber.FiberBundle.ring_number`.

```python
bundle.ring_number
```

***

##### Fiber bundle half.

Each fiber contains information on which half it is stored in, where zero is the top half and one is the bottom. This attribute can be accessed with `fiber.FiberBundle.half`.

```python
bundle.half
```

***

##### Fiber count.

This attribute is simply just the total number of fibers created within the bundle. It can be useful to know this value for future planning. The attribute can be accessed with `fiber.FiberBundle.count`.

```python
bundle.count
```

***

##### Fiber drawings. 

This attribute stores the drawings of each fiber in the bundle. These drawings are made solely for the purpose of plotting the fibers if you want to visualize the bundle. The attribute can be accessed with `fiber.FiberBundle.drawings`.

```python
bundle.drawings
```

***

### Visualizing the fiber bundle.

The fiber bundle is defined and we can extract plenty of information out of it, but let's say we want to see what it looks like. Luckily, there is a predefined function for this. Using `fiber.FiberBundle.plot(fibers = True, centers = False, ax = None, figsize = (10, 10), scatter_size = 1, im = Laser(), lc = "b", show = False, save = False, name = "", *args, **kwargs)`, you are able to plot the fiber bundle. This function seems quite complicated with lots of parameters, however it's actually quite simple in its execution.

Let's walk through all of the parameters in order: `fibers`, default behaviour shows all the individual fiber drawings; `centers`, default behaviour hides all of the centers of the fibers in the plot; `ax`, this parameter can take in an existing `matplotlib.pyplot.Axes` axes object but is optional because the function will create one if none is provided; `figsize`, a tuple that defines the width and height (in inches) of the generated plot; `scatter_size`, defines the size of the dots used to plot the centers of each fiber if `centers = True`; `im`, takes in the image of a laser beam if you'd like to overlay the position of the incident laser beam on the fiber bundle plot; `lc`, the color of the plot for laser beam; `show`, tells the function whether or not to run `matplotlib.pyplot.show()` for the axes object; `save`, tells the function whether or not to run `matplotlib.pyplot.savefig(name)` to save the figure to an image file; `name`, provides the name of the saved image file; `*args`, any arbitrary arguments that are applicable to a `matplotlib.pyplot.Axes.plot(*args, **kwargs)` function call; `**kwargs`, same as `*args` but for keyword arguments.

An example of a plotted fiber bundle is shown below.

```python
bundle.plot(figsize = (2, 2), show = True, c = "k")
```

***

### Modifying the fiber bundle.

##### Dropping fiber rings.

What about making modifications to a fiber bundle that has already been instantiated? This can be done in only a couple of ways. The first way is by dropping certain rings in the fiber bundle. Individual rings within the bundle can be dropped using their index, or a range of rings can be drapped by giving a start ring and end ring to drop from --- where the end is exclusive. The function to do this is `fiber.FiberBundle.drop_rings(i)`, where `i` is the index to drop or a tuple containing the range of indexes to drop.

```python
bundle.drop_rings(0)
bundle.plot(figsize = (2, 2), show = True, c = "k")
```

***

##### Resetting the bundle.

It can be seen that dropping the ring with `i = 0` removes the two innermost fibers in the bundle (or the zeroth ring). Now, let's say I'd like to drop multiple rows at once. I can do this by giving the function a `tuple` object with the range of rows that I'd like it to get rid of.

```python
bundle.drop_rings((0, 4))
bundle.plot(figsize = (2, 2), show = True, c = "k")
```

We run into a problem trying to immediately drop rings after previously modifying the bundle. In this case, it'd be useful to reset the bundle before making additional modifications. There is a function that will reset the bundle to its original configuration. Using `fiber.FiberBundle.reset()` sets the bundle back to default. Then we can freely modify it again.

```python
bundle.reset()
bundle.drop_rings((0, 3))
bundle.plot(figsize = (2, 2), show = True, c = "k")
```

### Making a Laser beam.

The second class that we'll use in these simulations is contained within the `laser.py` package. The class in this package is `laser.Laser(x = np.array([None]), y = np.array([None]), P = np.array([None]), xyz = np.array([]), res = 1506, waist = 1, power = 100, chop_waist = True, chop_percent = 1, power_noise = False)`. This class is used to instantiate a laser beam. It takes many parameters: x points of single rays that is left blank by default, `x`; y points of single rays that is left by default, `y`; power values that correspond with rays defined by their x and y position, `P`; a mesh grid that contains all of the x, y, and P points in a manner that allows for normalization and plotting left empty by default, `xyz`; square root of the number of points used to make the laser beam, `res`; the waist of the laser beam, `waist`; the total power of the beam when all the rays are summed, `power`; whether or not to chop the beam at its waist when building it, `chop_waist`; the percentage of the beam's waist at which to chop it, `chop_percent`; whether or not to add noise to the individual powers of each ray, `power_noise`.


```python
from laser import Laser # importing the Laser object from laser.py
l = Laser(waist = 3e-4, power = 90) # instantiating the laser beam
```

***

##### Visualizing the beam.

This laser beam can be plotted in a couple of different ways. These methods are `laser.Laser.plot_surface(ax = None, figsize = (10, 10), alpha = 1, cmap = cm.coolwarm, antialiased = False, show = False, save = False, name = "", *args, **kwargs)`, which uses the same arguments as the above plot function, and the less significant `laser.Laser.imshow(ax = None, dimensions = (1, 1), figsize = (10, 10), cmap = cm.coolwarm, save = False, origin = "lower", extent = [],  name = "", show = False)`. These are shown below.

```python
l.plot_surface(figsize = (2, 2), show = True)
l.imshow(figsize = (2, 2), show = True)
```

***

##### Showing the beam on the fiber bundle.

Now let's visualize the beam as it would hit on the fiber bundle.

```python
bundle.plot(im = l, c = "k", show = True, figsize = (2, 2))
```
