```python
import numpy as np
import matplotlib.pyplot as plt
r = 10
theta = np.linspace(0, 2 * np.pi, 1000)
plt.plot(r * np.cos(theta), r * np.sin(theta))
bundle_area = np.pi * r ** 2
fibers = 15
fiber_area_estimate = bundle_area / fibers
fiber_r = np.sqrt(fiber_area_estimate / np.pi)
n_of_fibers = int(r / fiber_r)
radius_to_draw = np.linspace(0, r - fiber_r, n_of_fibers)
if radius_to_draw[-1] + 2 * fiber_r > r:
    radius_to_draw = radius_to_draw[:-1]
arc_length = np.pi * radius_to_draw
n_fibers_per_ring = (arc_length / fiber_r + 1).astype(int)
# plt.plot(fiber_r * np.cos(theta) - (r - (fiber_r + spacing)) * np.cos(3 * np.pi / 2), fiber_r * np.sin(theta) - (r - (fiber_r + spacing)) * np.sin(3 * np.pi / 2))
for i in range(0, int(len(n_fibers_per_ring) / 2 + 1)):
# for i in range(0, 2):
    thetas = np.linspace(0, np.pi, n_fibers_per_ring[i])
    for t in thetas:
        plt.plot(fiber_r * np.cos(theta) - ((radius_to_draw[i] + fiber_r * i) * np.cos(t)), fiber_r * np.sin(theta) - ((radius_to_draw[i] + fiber_r * i) * np.sin(t)))
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import numpy as np
from scipy.stats import multivariate_normal
from scipy.optimize import curve_fit
from joblib import Parallel, delayed
from multiprocessing import Pool
import time
import itertools

p = 1000
waist = 3
xx, yy = np.mgrid[-(waist):(waist):(p * 1j), -(waist):(waist):(p * 1j)]

# Need an (N, 2) array of (x, y) pairs.
z = np.exp(-2 * (xx ** 2 + yy ** 2) ** 2 / (waist ** 2)) * np.abs(np.random.normal(-waist / 2, waist / 2, size = (p, p)))
z *= 200 / np.sum(z)
z = z.reshape(xx.shape)
# print(x.shape)
# print(x)
# print(y)
# print(z)
# print(z.shape)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xx, yy, z, cmap = cm.coolwarm, antialiased = False)
xy = np.column_stack([xx.flat, yy.flat])
mu = np.array([np.mean(xy[:, 0]), np.mean(xy[:, 1])])
# continue here where I left off
sigma = np.array([waist / 2, waist / 2])
covariance = np.diag(sigma ** 2)
z_fit = multivariate_normal.pdf(xy, mean=mu, cov=covariance)
z_fit *= np.max(z) / np.max(z_fit)
z_fit = z_fit.reshape(xx.shape)
ax.plot_wireframe(xx, yy, z_fit, alpha = 0.1, color = "k")
# print(z.shape)
# z = gaussian_bundle(x, y, *popt)
# print(z.shape)
# ax.plot_surface(x, y, )
#ax.plot_wireframe(x,y,z)
plt.show()
plt.imshow(z, norm = cm.colors.PowerNorm(gamma=0.25), cmap = cm.coolwarm, origin = "lower", extent = [np.min(xx), np.max(xx), np.min(yy), np.max(yy)])
plt.show()
plt.imshow(z_fit, norm = cm.colors.PowerNorm(gamma=1), cmap = cm.coolwarm, origin = "lower", extent = [np.min(xx), np.max(xx), np.min(yy), np.max(yy)])
plt.show()
def make_coord(x, y, z):
    return (x, y, z)
# t1 = time.time()
# coords = []
# with Parallel(n_jobs = -1) as parallel:
#     coords = np.array(parallel(delayed(make_coord)(xx[i][j], yy[i][j], z[i][j],) for i in range(len(xx)) for j in range(len(xx[i]))))
# t2 = time.time()
# print(t2 - t1)
# print(coords)
coords = []
t1 = time.time()
with Pool() as pool:
    for i in range(0, 1):
        for val in zip(xx[i], yy[i], z[i]):
            coords = list(pool.map(make_coord, args = (val[0], val[1], val[2]), chunksize = (1024 ** 3)))
t2 = time.time()
print(t2 - t1)
print(coords)
```
