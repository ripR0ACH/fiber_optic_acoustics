```python
import sys
sys.path.append("../../lhillber/brownian/src")
import os
from nptdms import TdmsFile
import scipy.stats as stats

def check_and_load_or_process(offset, *args,):
    results = process_folder(offset, *args)
    # Returns a list of args, and traces
    return results

def process_folder(offset, folder_name, data_col, num_traces, traces_per):
    results = []
    for i in range(num_traces):
        # print("Reading ", folder_name, str(i))
        # print("data_col ", data_col)
        for j in range(traces_per):
            result = process_file(folder_name, i, data_col, j, offset=offset)
            if result:
                results.append(result)
    return results

def process_file(folder_name, trace_num, data_col, trace_idx, offset):
    trace_num = trace_num + offset
    file_path = os.path.join(folder_name, "iter_" + str(trace_num) + ".tdms")
    series, args = read_tdms_file(file_path, data_col, trace_idx)

    return {
        "series": series,
        "args": args
    }
def read_tdms_file(file_path, data_col, trace_idx):
    tdms_file = TdmsFile.read(file_path)
    sample_rate = tdms_file["main"].properties.get("r", None)
    # print("Sample rate is " + str(sample_rate))
    series = tdms_file["main"][data_col + "_" + str(trace_idx)]
    track_len = len(series.data)
    config_args = {
        "sampling_rate": sample_rate,
        "track_len": track_len
    }

    return series[:], config_args

def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx

def bin_data(series, bin_size):
    # Ensuring the length of series is divisible by bin_size

    length = len(series) - len(series) % bin_size
    series = np.array(series[:length])
    ret = np.mean(series.reshape(-1, bin_size), axis=1)
    return ret

traces = check_and_load_or_process(176, "../data/sample_noise/", "X", 1, 5)
times = np.arange(0, len(traces[0]['series'])) * (1 / TdmsFile.read("../data/sample_noise/iter_176.tdms")["main"].properties.get("r", None))
print(TdmsFile.read("../data/sample_noise/iter_176.tdms")["main"].properties.get("r", None))
plt.plot(times[:find_nearest(times, 1e-4)], traces[0]["series"][:find_nearest(times, 1e-4)])
plt.plot(np.linspace(0, 0.0001, 100), bin_data(traces[0]["series"][:find_nearest(times, 1e-4)], 200))
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import periodogram
maxes = np.array([])
data = np.array([[]])
for i in np.linspace(5e-5, 1.25e-4, 50):
    with open("8fibers_" + str(i) + "beam_waist_with_cladding.csv") as f:
        file = csv.reader(f)
        file.__next__()
        data = np.append(data, np.array([*file], dtype = float))
        f.close()
for i in np.linspace(1.25e-4, 3e-4, 50):
    with open("8fibers_" + str(i) + "beam_waist_with_cladding.csv") as f:
        file = csv.reader(f)
        file.__next__()
        data = np.append(data, np.array([*file], dtype = float))
        f.close()
for i in np.linspace(3e-4, 1e-3, 50):
    with open("128fibers_" + str(i) + "beam_waist_with_cladding.csv") as f:
        file = csv.reader(f)
        file.__next__()
        data = np.append(data, np.array([*file], dtype = float))
        f.close()
data = data.reshape(150, -1, 3)
maxes = np.array([np.max(x) for x in data])
scale = 5 / np.max(maxes)
data[:, :, 1] = data[:, :, 1] * scale
for d in data:
    plt.plot(d[:, 0], d[:, 1])
plt.title("Signal vs. Time")
plt.ylabel("Signal (V)")
plt.xlabel("Time (s)")
plt.show()
x = np.append(np.append(np.linspace(5e-5, 1.25e-4), np.linspace(1.25e-4, 3e-4, 50)), np.linspace(3e-4, 1e-3, 50))
maxes *= scale
plt.axvline(1.25e-4, c = "k", label = "cover cladding", linestyle = "dashed")
plt.plot(x, maxes, label = "max vs. waist")
plt.xlabel("Beam Waist (m)")
plt.ylabel("Max Signal (V)")
plt.axvline(x[find_nearest(maxes, np.max(maxes))], c = "r", label = "max voltage", linestyle = "dashed")
print(x[find_nearest(maxes, np.max(maxes))])
plt.title("Max Signal vs. Beam Waist")
plt.legend()
plt.show()

plt.plot(data[find_nearest(maxes, np.max(maxes))][:, 0], data[find_nearest(maxes, np.max(maxes))][:, 1])
plt.xlabel("Time (s)")
plt.ylabel("Signal (V)")
plt.title("Best Signal Trace")
plt.savefig("signal.png")
plt.show()

plt.plot(data[find_nearest(maxes, np.max(maxes))][:, 0], bin_data(traces[0]["series"][:find_nearest(times, 1e-4)], 200))
plt.xlabel("Time (s)")
plt.ylabel("Signal (V)")
plt.title("Mephisto Noise Trace")
plt.savefig("noise.png")
plt.show()

f, x = periodogram(data[find_nearest(maxes, np.max(maxes))][:, 1] + bin_data(traces[0]["series"][:find_nearest(times, 1e-4)], 200), 1e6, nfft = 100)
plt.loglog(f, x)
plt.ylim(1e-15, 1e-1)
plt.ylabel(r"PSD ($V / \sqrt{Hz}$)")
plt.xlabel(r"Frequency (Hz)")
plt.title("Simulation PSD")
plt.show()
f, sig = periodogram(data[find_nearest(maxes, np.max(maxes))][:, 1] , 1e6, nfft = 100)
f, noi = periodogram(bin_data(traces[0]["series"][:find_nearest(times, 1e-4)], 200), 1e6, nfft = 100)
snr = sum(sig) / sum(noi)
print(10 * np.log10(snr))
```
